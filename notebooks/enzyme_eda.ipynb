## Create Data Dir structure

1. Go through json
2. Make a folder for each sample
3. Put the protein structure file in that folder
4. Pickle the smile string into the folder

```python
dataset_path = "/Mounts/rbg-storage1/datasets/Enzymes/DLKcat/DeeplearningApproach/Data/database/Kcat_combination_0918_wildtype_mutant_sample_ids.json"
```

```python
processed_dir = "/Mounts/rbg-storage1/datasets/Enzymes/predicted_processed_dir/"
```

```python
structures_dir = "/Mounts/rbg-storage1/datasets/Enzymes/DLKcat/DeeplearningApproach/Data/database/Kcat_combination_0918_wildtype_mutant_structures/"
```

```python
data = json.load(open(dataset_path, "rb"))
```

```python
df_data = pd.DataFrame(data)
```

```python
df_data[df_data.duplicated(subset="sample_id")]
```

```python
c = 0
for sample_id in df_data[df_data.duplicated(subset="sample_id")]['sample_id'].values:
    kcats = df_data[df_data['sample_id'] == sample_id]['Value'].values
    if len(set(kcats)) != 1:
        print(kcats)
        c += 1
print(c)
```

```python
len([val for val in df_data["seq_id"].values if val not in [1026, 1027, 514, 515, 1028, 1029, 1030, 1031, 2211, 6242, 7205, 2132]])
```

```python
for sample in tqdm(data):
    sample_id = sample["sample_id"]
    dir_name = "s".join(sample_id.split("_")[-2:])
    path_to_dir = os.path.join(processed_dir, dir_name)
    if not os.path.exists(path_to_dir):
        os.mkdir(path_to_dir)
    path_to_smile = os.path.join(path_to_dir, dir_name + "_smile.pkl")
    pdb_file = os.path.join(structures_dir, "seq_id_" + str(sample["seq_id"]) + ".pdb")
    if not os.path.exists(path_to_smile):
        pickle.dump(sample["Smiles"], open(path_to_smile, "wb"))
    if os.path.exists(pdb_file) and not os.path.exists(os.path.join(path_to_dir, dir_name + ".pdb")):
        shutil.copy(pdb_file, os.path.join(path_to_dir, dir_name + ".pdb"))
```

```python
# remove directories without pdb files
```

```python
counter = 0
seqs_missing = set()
for sample in tqdm(data):
    sample_id = sample["sample_id"]
    dir_name = "s".join(sample_id.split("_")[-2:])
    path_to_dir = os.path.join(processed_dir, dir_name)
    pdb_file = os.path.join(structures_dir, "seq_id_" + str(sample["seq_id"]) + ".pdb")
    if not os.path.exists(pdb_file):
        if os.path.exists(path_to_dir):
            shutil.rmtree(path_to_dir)
        seqs_missing.add(sample["seq_id"])
        counter += 1
        
print(counter)
print(len(seqs_missing))
print(seqs_missing)
```
